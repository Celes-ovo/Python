````
#그냥 넣으면 오류가 난다.
'that's not going to work'
````

````
"this will work"
````

해결법 2 : " "를 안에 넣고 싶을 경우, 바깥에 ' ' 를 넣으면 실행 가능하다.

````
'She said, "What does she said??"'
````

해결법 3 : ' '와 " "를 둘 다 넣고 싶을 경우, 밑의 코드처럼 안에 또 둘러 줘야 한다.

````
 'She said, "That' + "'" + 's hard to read."'
````

#9. Keywords (Kwargs)

함수 호출의 변수(parameter) 이름에 값(value)을 할당해 줌   
파이썬이 난 이 단어를 사용하겠다고 예약해 놓은 것

각각의 값들은 comma(,) 또는 빈칸(space)으로 구분한다.

```기본 seperator는 빈칸이다.```

````
print('a', 'b', 'c', flush = '.')
````

```여기에서는 seperator를 sep = ','를 이용해 빈칸과 ,로 바꿔주었다.```

````
print('a', 'b', 'c', sep=', ')
````

````
print('a', 'b', 'c', sep='@ ')     #다른 것도 되는군요?
````

```end = ''를 쳐 주면 새 줄로 넘어가지 않고 바로 다음 정보가 표시된다.```

````
print('a', 'b', 'c ', sep=', ', end='cat ') #\n
print('hocolate')
````
