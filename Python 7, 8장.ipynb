# 8.5 Aliasing

alias : 무언가의 또다른 이름 (?)

객체와 변수의 연관짓는 것을 참조(reference)라고 한다.   
하나 이상의 참조를 가진 객체는 한 개 이상의 이름을 갖게 되어서 객체가 aliased 되었다고 한다.

```
# When we modify the list using one of the variables,
# references through the other variable show the change as well:

celegans_phenotypes = ['Emb', 'Him', 'Unc', 'Lon', 'Dpy', 'Sma']
celegans_alias = celegans_phenotypes
celegans_phenotypes[5] = 'Lvl'
celegans_phenotypes
```

```
celegans_alias
```

어떤 두 variable x와 y가 있다. x를 aliasing한 게 y라고 하면   
x의 값이 바뀌게 될 경우 y의 값도 바뀌게 되고, 역 또한 성립한다.

```
x = ['Emb', 'Him', 'Unc', 'Lon', 'Dpy', 'Sma']
y = x
x[5] = 'Lvl'
x
```

```
y
```

```
y[3] = 'Lvl'
y
```

```
x
```

# 8.6 리스트의 요소 없애기

```
# del을 사용하면 된다.

x = ['Emb', 'Him', 'Unc', 'Lon', 'Dpy', 'Sma']
del x[-1]

x
```

`다른 예시`

```
def remove_last_item(L: list) -> list:
  """Return list L with the last item removed.
  Precondition: len(L) >= 0
  remove_last_item([1, 3, 2, 4])
  [1, 3, 2]
  """
  del L[-1]
```

```
celegans_markers = ['Emb', 'Him', 'Unc', 'Lon', 'Dpy', 'Lvl']
remove_last_item(celegans_markers)
celegans_markers
```
