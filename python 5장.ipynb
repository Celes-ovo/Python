```
print(45>30)
print(45>300)
print(25<56.0)
print(48<12)
```

```
print(65 == 64)
print(65 == 64.5)
print(65 == 65)
print(65 == 65.0)
print(65 == 65.1)
print(65 != 64)
print(65 != 64.5)
```

어떤 연산자라도 intergers와 floating-point numbers를 서로 비교할 수 있다.

```
#이미 알고 있는 값 대신 변수를 사용한 예

def is_positive(x: float) -> bool:
 """Return True iff x is positive.
 >>> is_positive(3)
 True
 >>> is_positive(-4.6)
 False
 """
 return x > 0

print(is_positive(3))
print(is_positive(-5))

"""
여기서 iff = if and only iff의 줄임말 = '정확히 언제' 라는 의미
"""
```

대부분, 이런 식으로 변수에 적용된다.

#6. 예상을 깨는 비교식

```
print(3 < 5 != True)
print(3 < 5 != False)
```

이 경우    
(3 < 5) and (5 != True)    
(3 < 5) and (5 != False)      
와 같다.

5는 True도 False도 아니니 뒤가 무조건 True가 된다.      
표현식의 나쁜 예라고 한다.   
1. 표현식은 보는 사람이 이해하기 쉬워야 한다.
2. 비교는 간단하게 하고, and를 쓰면 읽기 쉬워진다.

#7. 숫자와 문자열을 불 연산자와 함께 사용하기

파이썬은 0 또는 0.0을 False로 인식하고, 나머지 숫자를 True로 처리한다.

```
print(not 0)
print(not 1)
print(not 34.2)
print(not -8)
```

파이썬은 빈 문자열을 False로 취급하고 그렇지 않으면 True로 취급한다.

```
print(not '')
print(not 'bad')
```

None is also treated as False.

#8. Short circuit evaluation(or와 and expression에서)

* or expression의 첫 번째 operand가 True인 경우, 두 번째는 고려하지도 않는다.
* and expression의 첫 번째 operand가 False인 경우, 두 번째는 고려하지도 않는다.

#10. 실행할 명령문 고르기

if문을 사용하면 된다. (조건ㅇ문!)

어떤 if조건문이 있을 때, 그 조건에 부합하면 조건문에 해당하는 statement가 출력된다.

```
ph = 6.0
if ph < 7.0:
  print(ph, "is acidic.")

print("You should be careful with that!")
```

#11. elif

```
ph = float(input('Enter the pH level: '))

if ph < 7.0:
  print(ph, "is acidic.")

elif ph > 7.0:
  print(ph, "is basic.")
```

#12. Nested if statements(중첩 if문)

```
value = input('Enter the pH level: ')
if len(value) > 0:
  ph = float(value)

  if ph < 7.0:
    print(ph, "is acidic.")

  elif ph > 7.0:
    print(ph, "is basic.")

  else:
    print(ph, "is neutral.")
    
else:
  print("No pH value was given!")
```
