사용자가 그 module에서 어떤 것을 import할지 정확히 명시할 수 있다!

```
from math import sqrt, pi
print(sqrt(9))

radius = 5
print('circumference is', 2 * pi * radius)
```

#5. 모듈을 복구(restore)하기

import한 모듈의 variable이나 function의 값을 변경한 경우,   
shell을 재시작하고 모듈을 다시 import하면 원래 값이 복구된다.

셸을 재시작하지 않고도 importlib module의 reload 함수를 사용하면 사용자 정의 모듈을 다시 원상태로 돌릴 수 있다.    
importlib.reload 함수는 받았던 모듈을 다시 return한다.

```
import example
>>> example.x
2
>>> example.x = 7
>>> example.x
7


import importlib
>>> example = importlib.reload(example)
>>> example.x
2
```

#7. doctest module -> ?

docstring 안의 내용을 테스트 데이터로 쓸 수 있게 해 준다.

```
def div(x,y):

    """
    This function is to divide x into y.
   
    [examples]

    >>> div(1,2)
    0.5

    >>> div(4,3)
    1.3333333333333333

    >>> div(-1,4)
    -0.25
    """

    return x/y

import doctest
doctest.testmod()
```

docstring 안에 있는 함수 예시들을 다 한번에 돌려볼 수 있게 해 준다.

```
def div(x,y):

    """
    This function is to divide x into y.
   
    [examples]

    >>> div(1,2)
    0.5

    >>> div(4,3)
    1.3333333333333333

    >>> div(-1,4)
    -0.25
    """

    return x/y

import doctest
doctest.testmod()
```
